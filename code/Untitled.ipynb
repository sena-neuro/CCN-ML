```python
folder = "data/data_in/"+subj +"/"+subj+"_prior_video"
```

```python
android_data_p = np.asarray(ccn_io.read_file(folder + "_android.mat")[subj+"_prior_video_android"])
human_data_p   = np.asarray(ccn_io.read_file(folder +   "_human.mat")[subj+  "_prior_video_human"])
robot_data_p   = np.asarray(ccn_io.read_file(folder +   "_robot.mat")[subj+  "_prior_video_robot"])

and_average_over_trial = np.average(android_data_p,2)
hum_average_over_trial = np.average(human_data_p,2)
rob_average_over_trial = np.average(robot_data_p,2)
```

```python
import matplotlib.pyplot as plt
def plot_channel_range(a, b):
    
    print("RANGE -----------")
    for ch in range(a,b):
        plt.plot(rob_average_over_trial[ch])
    plt.show()

    for ch in range(a,b):
        plt.plot(and_average_over_trial[ch])
    plt.show()

    for ch in range(a,b):
        plt.plot(hum_average_over_trial[ch])
    plt.show()
```

```python
plot_channel_range(0,20)
print("RANGE -----------")
for ch in range(20):
    plt.plot(rob_average_over_trial[ch])
plt.show()

for ch in range(20):
    plt.plot(and_average_over_trial[ch])
plt.show()

for ch in range(20):
    plt.plot(hum_average_over_trial[ch])
plt.show()
```

```python
plot_channel_range(20,40)

print("RANGE -----------")
for ch in range(20,40):
    plt.plot(rob_average_over_trial[ch])
plt.show()

for ch in range(20,40):
    plt.plot(and_average_over_trial[ch])
plt.show()

for ch in range(20,40):
    plt.plot(hum_average_over_trial[ch])
plt.show()
```

```python

print("RANGE -----------")
for ch in range(40,60):
    plt.plot(rob_average_over_trial[ch])
plt.show()

for ch in range(40,60):
    plt.plot(and_average_over_trial[ch])
plt.show()

for ch in range(40,60):
    plt.plot(hum_average_over_trial[ch])
plt.show()
plot_channel_range(40,60)
```
